OpenAI API provides various finetune parameters for finetuning GPT models for various tasks such as text generation, text completion, question answering, sentiment analysis, and more. Here are some of the commonly used finetune parameters:


---


model: The ID of the base model to be used for finetuning (e.g., text-davinci-002).
prompt: The text prompt to be used for generating text or completing text.
temperature: The temperature parameter controls the randomness or creativity of the generated text.
max_tokens: The maximum number of tokens to be generated by the model.
stop: The stop sequence or token that indicates the end of the generated text.
n: The number of text completions to generate.
engine: The language model engine to be used for generating text (e.g., davinci, curie, babbage, etc.).
examples: The training examples to be used for finetuning the model.
validation: The validation examples to be used for evaluating the performance of the finetuned model.
batch_size: The number of training examples to be processed in a single batch.
learning_rate: The learning rate used for updating the model weights during training.
num_epochs: The number of epochs to train the model for.
logprobs: Whether to return the log probabilities of each generated token.
return_prompt: Whether to include the prompt in the generated text.

```python
from datetime import datetime
models = openai.Model.list()
models_id_list = []
for model in models['data']:
    models_id_list.append(model['id'])
# Define a function to extract the datetime from the model name
def extract_datetime(model_name):
    datetime_str = model_name.split('-')[-6:]
    datetime_str = '-'.join(datetime_str)
    try:
        return datetime.strptime(datetime_str, "%Y-%m-%d-%H-%M-%S")
    except ValueError:
        return None

# Filter out models with invalid datetime format and sort the remaining models
filtered_models = [model for model in models_id_list if extract_datetime(model)]
sorted_models = sorted(filtered_models, key=extract_datetime, reverse=True)

# Get the latest fine-tuned model if available
latest_model = sorted_models[0] if sorted_models else None

if latest_model:
    print("Latest Fine-tuned Model:", latest_model)
else:
    print("No valid fine-tuned models found.")
```

```python
import openai
import requests


def fine_tune_model(dataset, model_engine="davinci-2", num_epochs=10, batch_size=16):

    headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {openai.api_key}",
    }

    data = {
    "model": model_engine,
    "dataset": dataset,
    "num_epochs": num_epochs,
    "batch_size": batch_size
    }

    url = "https://api.openai.com/v1/fine-tunes"
    response = requests.post(url, headers=headers, json=data)

    # if response.status_code != 200:
    #     raise ValueError("Failed to fine-tune the model.")

    # Get the ID of the fine-tuned model
    model_id = response.json()["model_id"]
    return model_id
```
